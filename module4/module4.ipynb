```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def step_function(s):
    return np.array(s > 0).astype(int)
```

```python
x = np.arange(-10, 10, 0.1)
y = step_function(x)
```

```python
plt.plot(x, y)
plt.title('this is a proud title')
plt.show()
```

```python
def sigmoid(s):
    return 1/(1 + np.exp(-s))
```

```python
sigmoid(np.array([-2, -1, 0, 1, 2, 5]))
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.title('sigmoid function')
plt.show()
```
