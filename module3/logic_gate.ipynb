## AND gate
Let's make a simple AND gate.

```python
def and_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    th = 0.99

    if w1*x1 + w2*x2 > th:
        return 1
    else:
        return 0
```

```python
x1 = 0
x2 = 0
out = and_gate(x1, x2)
```

```python
print(f'The output of {x1} AND {x2} is {out}')
```

## LogicGate class

```python
class LogicGate:
    def __init__(self):
        self.w1 = None
        self.w2 = None
        self.th = None
        self.out = None
        self.x1 = self.x2 = None


    def print_output(self, gate):
        if gate == 'AND':
            print(f'{self.x1} AND {self.x2} is {self.out}')
        elif gate == 'OR':    
            print(f'{self.x1} OR {self.x2} is {self.out}')
    

    def or_gate(self, x1, x2):
        self.w1 = 0.5
        self.w2 = 0.5
        self.th = 0.0

        self.x1 = x1
        self.x2 = x2

        if x1*self.w1 + x2*self.w2 > self.th:
            self.out = 1
            return 1
        else:
            self.out = 0
            return 0        



    def and_gate(self, x1, x2):
        self.w1 = 0.5
        self.w2 = 0.5
        self.th = 0.99

        self.x1 = x1
        self.x2 = x2

        if x1*self.w1 + x2*self.w2 > self.th:
            self.out = 1
            return 1
        else:
            self.out = 0
            return 0
```

```python
logic_gate = LogicGate()
```

```python
logic_gate.and_gate(1, 0)
logic_gate.print_output('AND')
```

```python
logic_gate.or_gate(1, 0)
logic_gate.print_output('OR')
```
